## 9. Inheritance — Task
Create GraduateStudent.

```python
class graduate_student(student):
    def __init__(self, age, name, grad_flag):
        super().__init__( age, name)
        self.grad_flag=grad_flag
    def print_data(self):
        print("name", self.name, " age", self.age, "grad flag: ",self.grad_flag)
        
s2=graduate_student(24,"esraa", True)
s2.print_data()
```

## 10. Decorators — Task
Logger decorator.

```python
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__}")
        result=func(*args, **kwargs)
        print(f"{func.__name__} finished")
    return wrapper

@log_decorator
def greet(name):
    print("hello ", name)

greet("Alice")
```

## 11. Generators — Task
Countdown generator.

```python
def countdown(n):
    while n>0:
        yield n
        n-=1

for num in countdown(5):
    print(num)
```

## 13. Advanced Practice

```python
numbers=[1,2,3,4,5]
print(max(numbers))
print(min(numbers))
print(sum(numbers)/len(numbers))
#agent class
class Agent:
    def __init__(self, name, role):
        self.name=name
        self.role=role
    def introduce(self):
        print("my name is ", self.name)
        print("my role is ", self.role)
    
    def work(self):
        print(f"{self.name} is working")

agent1=Agent("Esraa", "AI engineer")
agent1.introduce()
agent1.work()


#generator divisable by 3
def divisable_by_3(limit):
    for i in range(1,limit+1):
        if i%3==0:
            yield i
for num in divisable_by_3(10):
    print(num)
```
